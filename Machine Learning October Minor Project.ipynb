```python
data['location'].value_counts()
```

```python
data['location']=data['location'].fillna(data.location.value_counts().index[0])
```

```python
data['rest_type'].value_counts()
```

```python
data['rest_type']=data['rest_type'].fillna(data.rest_type.value_counts().index[0])
```

```python
data['dish_liked'].value_counts()
```

```python
data['dish_liked']=data['dish_liked'].fillna(data.dish_liked.value_counts().index[0])
```
